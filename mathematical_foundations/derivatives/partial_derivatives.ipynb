```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# implement f(x,y) numerically
def f(x, y):
    return x**2 + y**2
```

```python
# Estimation of the gradient
#approximation of the derivative of the function 
h = 0.00000001 #the step
x = 2
y = 3
dxy =((f(x+h, y)- f(x,y))/h)
dyx = ((f(x, y+h)- f(x,y))/h)
print(dxy)
print(dyx)
```

```python
x = np.arange(-6, 6, 0.25)
y = np.arange(-6, 6, 0.25)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contourf(X, Y, Z)
plt.colorbar()

plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x,y) = x² + y²")

plt.show()
```
