```python
import numpy as np
x = np.arange(25).reshape(5, 5)
print(x)
```

```python
'''Extract the diagonal without using numpy.diagonal
diagonal is elements are those where row index =column index '''
rows = np.array([0, 1, 2, 3, 4])
cols = np.array([0, 1, 2, 3, 4])
diagonal = x[rows, cols]
print(diagonal)

'''or we can do:
idx = np.arange(5)
diagonal = x[idx, idx]'''
```

```python
# extraxting all elements greater than 10 using boolean extracting 
result = x[x > 10]
print(result)
```

```python
rev_rows = x[:, ::-1]
print(rev_rows)
```
