```python
import math
import numpy as np
```

```python
#from a list
a = np.array([1, 2, 3, 4, 5])
print(a.dtype, a.shape)

# using zeros
b = np.zeros(5)
print(b.dtype, b.shape)

# using ones
c = np.ones(5)
print(c.dtype, c.shape)

# using arange
d = np.arange(0, 10, 2)  
print(d.dtype, d.shape)

# using linspace
e = np.linspace(0, 1, 5) 
print(e.dtype, e.shape)
```
