```python
import numpy as np
#Same arrays as before but with different types to check memory usage
#from a list
a = np.array([1, 2, 3, 4, 5], dtype=np.int32)
print(a.nbytes)

# using zeros
b = np.zeros(5, dtype = np.float64)
print(b.nbytes)

# using ones
c = np.ones(5, dtype=bool)
print(c.nbytes)

# using arange
d = np.arange(0, 10, 2, dtype=np.int16)  
print(d.nbytes)

# using linspace
e = np.linspace(0, 1, 5, dtype=np.float32) 
print(e.nbytes)

b = np.zeros(5, dtype=np.float64)
e = np.linspace(0, 1, 5, dtype=np.float32)

result = b * e

print(result.dtype)
```
